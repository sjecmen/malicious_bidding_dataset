```python
##################################################################
#
#                  Anonymize the grouping
#
##################################################################
import csv
import random
import string
import time

groups = {}
with open('unanonymize/group.tsv') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter='\t')
    for row in csv_reader:
        groups[row[0]] = row[1]

list_of_id_map = {}
with open('unanonymize/mapping.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        list_of_id_map[row[0]] = row[1]

profiles = {}
with open('unanonymize/profiles.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        profiles[row[0]] = (row[1], row[2])

anonymized_groups = {}
with open('anonymize/anonymize_group.csv', 'w', newline='') as csvfile:
    column_names = ['Group', 'Members']
    csv_writer = csv.DictWriter(csvfile, fieldnames=column_names)
    csv_writer.writeheader()
    for idx in groups:
        members = ''
        names = ''
        count = 0
        for student in profiles:
            if profiles[student][0] in groups[idx] and profiles[student][1] in groups[idx]:
                if student in list_of_id_map:
                    members += list_of_id_map[student] + ', '
                names += profiles[student][0] + ' ' + profiles[student][1] + ', '
                count += 1
        csv_writer.writerow({'Group': idx, 'Members': members})
        print(names)
```

```python
##################################################################
#
#                 Un-anonymize the malicious group
#
##################################################################
import csv
import random
import string
import time

list_of_column_names = []
list_of_students = {}

with open('../steven/new_groups_anonymized.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        list_of_column_names = row
        break
    for row in csv_reader:
        list_of_students[row[0]] = row[1:]

list_of_id_map = {}
with open('unanonymize/mapping.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        list_of_id_map[row[1]] = row[0]


with open('../steven/new_groups.csv', 'w', newline='') as csvfile:
    csv_writer = csv.DictWriter(csvfile, fieldnames=list_of_column_names)
    csv_writer.writeheader()
    for key in list_of_students:
        one_row = {}
        one_row['name'] = list_of_id_map[key]
        for c_id, column in enumerate(list_of_column_names[1:]):
            one_row[column] = list_of_students[key][c_id]
        csv_writer.writerow(one_row)
```

```python
##################################################################
#
#                  Anonymize the biddings
#
##################################################################
import csv
import random
import string
import time

# Read in email to id map
list_of_id_map = {}
with open('unanonymize/mapping.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        list_of_id_map[row[0]] = row[1]

# Read in honest bidding
for name in ['honest_bidding', 'malicious_bidding']:
    bidding = {}
    with open(f'unanonymize/{name}.csv') as csvfile:
        csv_reader = csv.reader(csvfile, delimiter=',')
        header = next(csv_reader)
        email_idx = header.index('RecipientEmail')
        start_idx = header.index('Q3_1')
        new_header = [header[email_idx]] + header[start_idx:]
        header2 = next(csv_reader)
        new_header2 = [header2[email_idx]] + header2[start_idx:]
        next(csv_reader)
        for row in csv_reader:
            bidding[row[email_idx]] = row[start_idx:]

    with open(f'anonymize/anonymize_{name}.csv', 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile, delimiter=',')
        csv_writer.writerow(new_header)
        csv_writer.writerow(new_header2)
        for email, row in bidding.items():
            anon = list_of_id_map[email]
            csv_writer.writerow([anon] + row)
```

```python
##################################################################
#
#                  Anonymize the setup (from steven/new_groups.csv)
#
##################################################################
import csv
import random
import string
import time

# Read in email to id map
list_of_id_map = {}
with open('unanonymize/mapping.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        list_of_id_map[row[0]] = row[1]

# Read in and write setup
with open(f'unanonymize/setup.csv') as csvfile, open(f'anonymize/anonymize_setup.csv', 'w', newline='') as outfile:
    csv_reader = csv.reader(csvfile, delimiter=',')
    header = next(csv_reader)
    email_idx = 0
    csv_writer = csv.writer(outfile, delimiter=',')
    csv_writer.writerow(header)
    for row in csv_reader:
        email = row[email_idx]
        anon = list_of_id_map[email]
        csv_writer.writerow([anon] + row[email_idx+1:])
```
